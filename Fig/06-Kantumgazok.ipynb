```python
x=linspace(0,4,1000)
figure(figsize=(12,6))
fs=30
plot(x,1/(exp(10*(x-2))+1),lw=3)
grid()
ylim(0,1.2);
aprop=dict(color='k',width=.003)
yticks(linspace(0,1,3),(list(map(lambda x:r'$'+str(x)+r'$',linspace(0,1,3)))),fontsize=fs)
xticks(linspace(0,4,5),[r'$0$','',r'$\mu$',''],fontsize=fs)
xlabel(r'$E$',fontsize=fs)
ylabel(r'$f(E)$',fontsize=fs)
arrow(2,0.5,3/10,0,**aprop)
arrow(2,0.5,-3/10,0,**aprop)
text(1.8,0.6,r'$k_\mathrm{B}T$',fontsize=fs);
```
